```python
# Dependencies and Setup
import pandas as pd
import os

# File to Load (Remember to Change These)
school_data_to_load = "Resources\schools_complete.csv"
student_data_to_load = "Resources\students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
schools_unique = df["school_name"].unique()
total_schools= len(schools_unique)
total_students = df["student_name"].count()
total_budget = "$" + str(school_data["budget"].sum())
average_math = round(df["math_score"].mean(), 2)
average_reading = round(df["reading_score"].mean(), 2)
overall_passing = round(((average_reading + average_math)/2), 2)
reading_above_70 = df[df["reading_score"] >= 70]
reading_passing =round(((reading_above_70["reading_score"].count())/(total_students)*100), 2)
math_above_70 = df[df["math_score"] >= 70]
math_passing = round(((math_above_70["math_score"].count())/(total_students)*100), 2)
data_for_district_summary = [{"Number of Unique Schools": total_schools, "Number of Students": total_students,\
                              "District Budget": total_budget, "Average Math Score": average_math,
                             "Average Reading Score": average_reading,"% Passing Math": math_passing,\
                             "% Passing Reading": reading_passing, "Overall Passing": overall_passing}]
pd.DataFrame(data_for_district_summary)
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
school_summary_df = pd.DataFrame(school_data).rename(columns= {"school_name": "School Name", "type":"School Type",\
                                           "size": "School Size", "budget": "School Budget"}).sort_values(by=["School Name"]).set_index("School Name")
per_student_budget = (school_summary_df["School Budget"])/(school_summary_df["School Size"])
school_summary_df["Per Student Budget"] = per_student_budget
schools_grouped_mean = df.groupby(["school_name"]).mean()
reading_average = pd.DataFrame(schools_grouped_mean["reading_score"])
school_summary_df["Reading Score Average"] = round(reading_average,2)
math_average = pd.DataFrame(schools_grouped_mean["math_score"])
school_summary_df["Math Score Average"] = round(math_average, 2)
df_renamed = df.rename(columns={"school_name": "School Name"})
df_school= df_renamed.set_index("School Name")
df_school_passing_reading = df_school[df_school["reading_score"] >= 70]
reading_grouped = df_school_passing_reading.groupby("School Name").count()
reading_grouped_isolated = pd.DataFrame(reading_grouped["reading_score"])
percent_pass_reading = (reading_grouped_isolated["reading_score"])/(school_summary_df["School Size"])*100
per_pass_read_df = pd.DataFrame(percent_pass_reading)
school_summary_df["% Passing Reading"] = round(per_pass_read_df, 2)
df_school_passing_math = df_school[df_school["math_score"] >= 70]
math_grouped = df_school_passing_math.groupby("School Name").count()
math_grouped_isolated = pd.DataFrame(math_grouped["math_score"])
percent_pass_math = (math_grouped_isolated["math_score"])/(school_summary_df["School Size"])*100
per_pass_math_df = pd.DataFrame(percent_pass_math)
school_summary_df["% Passing Math"] = round(per_pass_math_df, 2)
overall_passing = round(((school_summary_df["% Passing Reading"] + school_summary_df["% Passing Math"])/2), 2)
school_summary_df["Overall Passing %"] = overall_passing
all_school_summary_df = pd.DataFrame(school_summary_df)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
summdf= all_school_summary_df
reading_grouped_isolated_df = pd.DataFrame(reading_grouped_isolated)
math_grouped_isolated_df = pd.DataFrame(math_grouped_isolated)
reading_math_df = reading_grouped_isolated_df.merge(math_grouped_isolated_df, left_index= True, right_index= True, how="outer")
all_numerical_df = summdf.merge(reading_math_df, left_index = True, right_index = True, how="outer")
```

```python
spending_bins = [0, 585, 615, 645, 675]
spending_bin = ["<$585", "$585-615", "$615-645", "$645-675"]

complete_spending_reading_df = all_numerical_df
complete_spending_math_df = all_numerical_df

complete_spending_reading_df["Spending Range"] = pd.cut(complete_spending_reading_df["Per Student Budget"], bins=spending_bins, labels=spending_bin)
complete_spending_reading_grouped = complete_spending_reading_df.groupby("Spending Range").sum()
complete_spending_reading_passing = pd.DataFrame(round((((complete_spending_reading_grouped["reading_score"])/complete_spending_reading_grouped["School Size"])*100),2))

complete_spending_math_df["Spending Range"] = pd.cut(complete_spending_math_df["Per Student Budget"], bins=spending_bins, labels=spending_bin)
complete_spending_math_grouped = complete_spending_math_df.groupby("Spending Range").sum()
complete_spending_math_passing = pd.DataFrame(round((((complete_spending_math_grouped["math_score"])/complete_spending_math_grouped["School Size"])*100),2))

complete_spending_math_reading_passing = complete_spending_reading_passing.merge(complete_spending_math_passing, left_index = True, right_index = True, how="outer")
complete_spending_math_reading_passing_renamed = complete_spending_math_reading_passing.rename(columns={"0_x": "% Passing Reading", "0_y": "% Passing Math"})
overall_passing_spending = round(((complete_spending_math_reading_passing_renamed["% Passing Reading"] + complete_spending_math_reading_passing_renamed["% Passing Math"])/2), 2)
complete_spending_math_reading_passing_renamed["Overall Passing %"] = overall_passing_spending
```

```python
spendingdf = all_school_summary_df
spendingdf["Spending Range"] = pd.cut(spendingdf["Per Student Budget"], bins=spending_bins, labels=spending_bin)
spending_df_groupby = round((spendingdf.groupby("Spending Range").mean()), 2)
spending_df_groupby_mean = spending_df_groupby.drop(columns= ["School ID", "School Size", "School Budget",\
                                                             "Per Student Budget", "% Passing Reading",\
                                                             "% Passing Math", "Overall Passing %"])
round(spending_df_groupby_mean,2)
final_school_spending_df = spending_df_groupby_mean.merge(complete_spending_math_reading_passing_renamed, left_index = True, right_index = True, how="outer")
```
